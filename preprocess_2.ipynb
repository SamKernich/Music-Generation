```python
import os
import music21 as m21

MIDI_DATASET_PATH = "Dataset_classical/bach"
```

```python
# durations are expressed in quarter length
ACCEPTABLE_DURATIONS = [
    0.25, # 16th note
    0.5, # 8th note
    0.75,
    1.0, # quarter note
    1.25,
    1.5,
    1.75,
    2, # half note
    2.25,
    2.5,
    2.75,
    3,
    3.25,
    3.5,
    3.75,
    4 # whole note
]
```

```python
def load_songs_in_midi(dataset_path):
    """Loads all kern pieces in dataset using music21.

    :param dataset_path (str): Path to dataset
    :return songs (list of m21 streams): List containing all pieces
    """
    songs = []

    # go through all the files in dataset and load them with music21
    for path, subdirs, files in os.walk(dataset_path):
        for file in files:

            # consider only kern files
            if file[-3:] == "mid":
                song = m21.converter.parse(os.path.join(path, file))
                songs.append(song)
    return songs
```

```python
def has_acceptable_durations(song, acceptable_durations):
    """Boolean routine that returns True if piece has all acceptable duration, False otherwise.

    :param song (m21 stream):
    :param acceptable_durations (list): List of acceptable duration in quarter length
    :return (bool):
    """
    stream = m21.stream.Stream()
    usable_note_dur = []
    false_cnt = 0
    for note in song.flatten().notesAndRests:
        #print(note.duration.quarterLength)
        if note.duration.quarterLength not in acceptable_durations:
            false_cnt += 1
            continue
        if isinstance(note, m21.note.Note):
            stream.append(note)
            usable_note_dur.append(note.duration.quarterLength)
    print(usable_note_dur)
    stream.show('midi')
    if false_cnt > 0:
        return False
    else:
        return True
```

```python
songs = load_songs_in_midi(MIDI_DATASET_PATH)
print(f"Loaded {len(songs)} songs.")
```

```python
song = songs[2]
print(f"Has acceptable duration? {has_acceptable_durations(song, ACCEPTABLE_DURATIONS)}")
```
